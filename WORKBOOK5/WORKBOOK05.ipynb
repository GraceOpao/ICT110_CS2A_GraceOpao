```python
import numpy as np  # Import numpy library and assign an alias 'np'

# A. Create an ndarray object (A) with values from 1 to 100 in a nested Python List.
# You can choose the range from the given values (1-100).
A = np.array([[i for i in range(1, 101)]])  # Create a nested Python list containing values from 1 to 100,
                                            # then convert it into a numpy array using np.array().

# Accessing properties and methods of the ndarray object 'A':
print("Shape of A:", A.shape)  # Display the shape of array A using the 'shape' property.
print("Maximum value in A:", A.max())  # Find and print the maximum value in array A using the 'max()' method.
print("Sum of all values in A:", A.sum())  # Calculate and print the sum of all values in array A using the 'sum()' method.
print("Mean of all values in A:", A.mean())  # Calculate and print the mean of all values in array A using the 'mean()' method.

# Using numpy functions and passing the ndarray 'A':
print("Minimum value in A:", np.min(A))  # Find and print the minimum value in array A using the 'min()' function.

# B. Create a 5x5 integer-array filled with ones.
# Check its shape, size, and number of dimensions.
ones_array = np.ones((5, 5), dtype=int)  # Create a 5x5 array filled with ones using the 'ones()' function.
print("\nShape of ones_array:", ones_array.shape)  # Display the shape of ones_array.
print("Size of ones_array:", ones_array.size)  # Display the size (total number of elements) of ones_array.
print("Number of dimensions of ones_array:", ones_array.ndim)  # Display the number of dimensions of ones_array.

# C. Create an array of values from -5 to +5. Multiply each value by 2. Print the resulting array.
# Explanation: We use np.arange() to create an array of values from -5 to 5, then multiply each value by 2.
result_array = np.arange(-5, 6) * 2  # Create an array of values from -5 to 5 using np.arange() and multiply each value by 2.
print("\nResulting array after multiplication:", result_array)  # Print the resulting array.
```

```python
#A. A four-dimensional array of random integers in the range [0, 12] is produced using this code.
# It then retrieves the size, shape, and dimensions of the resultant array and prints it.


import numpy as np

# Create a four-dimensional array of random integers in the interval [0, 12]
arr = np.random.randint(0, 13, size=(3, 4, 5, 2))

# Print the resulting array
print("Resulting array:")
print(arr)

# Get the dimensions, shape, and size of the array
print("\nDimensions:", arr.ndim)
print("Shape:", arr.shape)
print("Size:", arr.size)

# B. This code gets the maximum number for each dimension of the array and displays it.

# Get the maximum number for each dimension
max_values = [arr.max(axis=i) for i in range(arr.ndim)]
print("\nMaximum number for each dimension:")
for i, max_val in enumerate(max_values):
    print(f"Dimension {i+1}: {max_val}")

# C. This code multiplies each element of the array by 4.

# Multiply each array with 4
result = arr * 4

# An explanation of the process of multiplication
# The array's values are essentially scaled up by a factor of four by multiplying each element by 4.
# This operation is performed element-wise, which means that the value of each element is multiplied by 4 on its own.

```
